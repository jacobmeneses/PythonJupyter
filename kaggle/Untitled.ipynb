```python
def simple_parse(some_str):
    replaced = some_str.replace('$','')
    return float(replaced)

def range_parse(some_str):
    replaced = some_str.replace('$','')
    splitted = replaced.split('-')
    if len(splitted) == 2:
        lower = float(splitted[0].strip())
        upper = float(splitted[1].strip())
        return (lower + upper) / 2.0
    raise RuntimeError from None
    
parse_functions = [ simple_parse, range_parse ]

def process_cost(cost_str):
    parsed = float("nan")
    for fn in parse_functions:
        try:
            parsed = fn(cost_str)
            break
        except:
            parsed = float("nan")
    return parsed    
    
abs_df_2 = abs_df.assign(real_cost=lambda x: x['cost'].apply(process_cost))
abs_df_2
#abs_df_2['cost'][1]
```
